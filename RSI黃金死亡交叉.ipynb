```python
from workshop import get_price
from workshop import MA, RSI, crossover
from account import Account
```

```python
data = get_price(2330, 2317)
```

```python
RSI(2330, 5)
RSI(2330, 15)

# 為了驗證
crossover(2330, 'RSI5', 'RSI15')
```

```python
a = Account()
pre_rsi5 = 0
pre_rsi15 = 0
rsi_crossover = False
rsi_crossunder = False
for date, price in data.iterrows():
    print('-------', date, 'start', '-------')
    a.handle_order(price)
    
    # 2330 RSI crossover
    c1 = price[2330, 'RSI5'] > price[2330, 'RSI15']
    c2 = pre_rsi5 < pre_rsi15
    rsi_crossover = c1 and c2
                
    # 2330 RSI crossunder
    c1 = price[2330, 'RSI5'] < price[2330, 'RSI15']
    c2 = pre_rsi5 > pre_rsi15
    rsi_crossunder = c1 and c2
            
    pre_rsi5, pre_rsi15 = price[2330, 'RSI5'], price[2330, 'RSI15']
    
    # 進出場判斷
    if a.is_empty() and rsi_crossover:
        print('crossover~~~~~')
        a.buy(2330)
    
    if not a.is_empty() and rsi_crossunder:
        print('crossunder')
        a.sell()
    
    if price[2330, 'RSI5xRSI15']:
        print('crossoverrrrrrr')
    print(f'-------- {date} end --------\n\n')
    pass
a.sell()
a.handle_order(price)
```
